````python
PATH = '/content/gdrive/My Drive/Github/MLZoomcamp2021/data/'
model_file = 'model1.bin'
dv_file = 'dv.bin'

customer = {"contract": "two_year", 
            "tenure": 12, 
            "monthlycharges": 19.7}

with open(PATH + model_file, 'rb') as f:
    model = pickle.load(f)

with open(PATH + dv_file, 'rb') as f:
    dv = pickle.load(f)

X = dv.transform(customer)
prediction = model.predict_proba(X)[0,1]
print(f'Probability of churning: {prediction:.3f}')
````
